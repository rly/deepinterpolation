```python
path = r"C:\Users\Ryan\Documents\SpikeSorting\deepinterpolation\data\PAIRED_KAMPFF\c14\c14_npx_raw.bin"
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npx_channels = 384
npx_recording = np.memmap(path, mode = 'r', dtype=np.int16, order = 'C')
npx_samples = int(len(npx_recording)/npx_channels)
npx_recording = npx_recording.reshape((npx_channels, npx_samples), order = 'F')
```

```python
di_output = np.random.rand(100000) * 40 - 20
channel_index = 160

end_t = 100000
plt.plot(npx_recording[channel_index, :end_t])
# plt.plot(di_output[channel_index, :end_t])
plt.show()
```

```python
# probably no spikes in this time window. use to compute standard deviation
t_no_spikes = range(105000, 110000)
plt.plot(npx_recording[channel_index, t_no_spikes])
plt.plot(di_output[channel_index, t_no_spikes])
plt.show()
```

```python
plt.plot(di_output[channel_index, t_no_spikes] - npx_recording[channel_index, t_no_spikes])
plt.show()
```

```python
stdev = np.std(npx_recording[channel_index, t_no_spikes])
print("std before deep interpolation %0.2f" % stdev)
di_stdev = np.std(di_output[channel_index, t_no_spikes])
print("std after deep interpolation %0.2f" % di_stdev)
```

```python
# zoom in on one (likely) spike waveform
t_spike = range(103620, 103740)
plt.plot(npx_recording[channel_index, t_spike])
# plt.plot(di_output[channel_index, t_spike])
plt.show()
```

```python
plt.plot(di_output[channel_index, t_spike] - npx_recording[channel_index, t_spike])
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 5))
ax[0].imshow(npx_recording[:, t_spike], aspect='auto')  # stretches x axis
ax[1].imshow(npx_recording[:, t_spike], aspect='auto')
plt.show()
```
